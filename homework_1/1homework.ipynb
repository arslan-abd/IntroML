**Задание 9: В этом задании вам предстоит построить ваш первый график. Ответьте на вопрос: людей с каким образованием больше всего представленно в датасете? \

```python
# Write code here
# book = {}

# for l in df['education']:
#   print(l)
education_counts = df['education'].value_counts()
print(education_counts)
plt.figure(figsize=(12, 6))
education_counts.plot(kind='bar', color='skyblue')
plt.title('Распределение людей по уровню образования', fontsize=16)
plt.xlabel('Уровень образования', fontsize=12)
plt.ylabel('Количество людей', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
most_common_education = education_counts.index[0]
most_common_count = education_counts.iloc[0]
print(f"\nБольше всего представлено людей с образованием: '{most_common_education}' ({most_common_count} человек)")
```

**Задание 10: А теперь давайте посмотрим на распределение возрастов. Сколько возрастных отрезков имеют больше 1600 значений? (Одним "отрезков" является один столбец при стандартных настройках seaborn/matplotlib)
P.S. Необязательное задание, хотя лучше сделать, если собираетесь сдавать до мягкого, я хоты бы посмотрю**

```python
# Write code here
print("ОСНОВНАЯ ИНФОРМАЦИЯ О ВОЗРАСТЕ:")
print("=" * 40)
print(f"Всего записей: {len(df)}")
print(f"Минимальный возраст: {df['age'].min()} лет")
print(f"Максимальный возраст: {df['age'].max()} лет")
print(f"Средний возраст: {df['age'].mean():.1f} лет")
print(f"Медианный возраст: {df['age'].median()} лет")
plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(df['age'], bins='auto', alpha=0.7, color='skyblue', edgecolor='black')

plt.title('Распределение возраста в датасете', fontsize=16)
plt.xlabel('Возраст', fontsize=12)
plt.ylabel('Количество людей', fontsize=12)
plt.grid(alpha=0.3)


plt.axhline(y=1600, color='red', linestyle='--', linewidth=2, label='Порог: 1600 человек')


bins_above_1600 = sum(n > 1600)
print(f"\nКоличество возрастных отрезков с >1600 значений: {bins_above_1600}")


print("\nБины с более чем 1600 значениями:")
for i, (bin_start, bin_end, count) in enumerate(zip(bins[:-1], bins[1:], n)):
    if count > 1600:
        print(f"Бин {i+1}: [{bin_start:.1f}-{bin_end:.1f}) лет - {count:.0f} человек")

plt.legend()
plt.tight_layout()
plt.show()
```
